## seekfeatures
This notebook demonstrates the use of ``seekfeatures`` using data HCEPD_100K.csv as example.
<br>

``seekfeatures`` is a module that extracts chemical features from input structure in the form of Simplified molecular-input line-entry system (**SMILES_str**). By using open-source package ``mordred``, this module returns more than 1800 chemical 1800 descriptors at high speed. 

<br>``seekfeatures`` contains three functions:<br>
> `seekfeatures.chem_features` extracts chemical information from molecular structures in the form of Smiles_str. <br>
>
> `seekfeatures.missing_value_list` returns a list of missing values with location in the input dataframe.<br>
>
> `seekfeatures.replace_missing` replaces all non-value entries(marked as specific mordred.error.Missing message in the         
                                dataframe with a value 0.

```python
from seekfeatures import *

# other modules needed for this demonstration
import pandas as pd
import numpy as np
```

<br>

* **`seekfeatures.chem_features`**

```python
# Use 100K data for demonstration
data = pd.read_csv('../Database/HCEPD_100K.csv') 
# Use first 50 data to illustrate the analysis
data50 = data.head(50)
data50.head()
```

```python
# input of ChemFeatures should be a list of string
features_df = chem_features(data50['SMILES_str'])
```

```python
features_df.head()
```

 <br>

* **`seekfeatures.missing_value_list`**<br>

Not all molecules have the same features as others, so there will be some non-value entries hidden in the dataframe.

```python
missing_list = missing_value_list(features_df)
missing_list[0:6]
```

* **`seekfeatures.replace_missing`**<br>

For these non-value entries, we assign them to 0, simply indicating that the molecule does not possess that specific feature.

```python
features_df_n = replace_missing(features_df)
```

```python
features_df_n.head()
```

```python
#Check if there is missing value after replacement
missing_list = missing_value_list(features_df_n)
missing_list
```
